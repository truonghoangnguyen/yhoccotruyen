```python
import asyncio
import os
import re
from slugify import slugify
from crawl4ai import AsyncWebCrawler, CrawlerRunConfig
config = CrawlerRunConfig(
    # e.g., first 30 items from Hacker News
    css_selector="#cscontentSize"
)

# # Danh sách URLs
# urls = [
#     {"url": "https://trungtamthuoc.com/duoc-lieu/actiso-58", "title": "atiso"},
#     {"url": "https://trungtamthuoc.com/duoc-lieu/bach-thuoc", "title": "bach-thuoc"},
#     # Thêm các URL khác...
# ]

def sanitize_filename(filename):
    """Loại bỏ ký tự không hợp lệ trong tên file"""
    return re.sub(r'[<>:"/\\|?*]', '_', filename)

async def crawl_and_save(url_data, crawler, folder):
    """Crawl một URL và save thành file markdown"""
    try:
        result = await crawler.arun(
            url=url_data["url"],
            config = config
        )

        if result.success:
            # Tạo tên file an toàn
            cleaned = re.sub(r'\s*\([^)]*\)', '', url_data["title"])
            filename = slugify(cleaned) + ".md"

            # Save markdown content
            with open(folder + "/" + filename, 'w', encoding='utf-8') as f:
                f.write(f"# {url_data['title']}\n\n")
                f.write(result.markdown)
                f.write(f"\n\nSource: {url_data['url']}")

            print(f"✅ Saved: {filename}")
        else:
            print(f"❌ Failed: {url_data['title']} - {result.error_message}")

    except Exception as e:
        print(f"❌ Error crawling {url_data['title']}: {str(e)}")

async def crawl_all():
    """Crawl tất cả URLs"""
    async with AsyncWebCrawler(verbose=True) as crawler:
        # Crawl tuần tự (tránh overload server)
        for url_data in url_list:
            await crawl_and_save(url_data, crawler, "trungtamthuoc.com")
            await asyncio.sleep(2)  # Delay 1 giây giữa các request

# Chạy
await crawl_all()
```
